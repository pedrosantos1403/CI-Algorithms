```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import LinearRegressionOneVariable as lr
```

```python
df = pd.read_csv("C:\\Users\\pedro\\OneDrive\\Documentos\\CEFETMG\\2023.2\\Inteligência Computacional\\CI-Algorithms\\Linear Regression\\One Variable\\data1.txt", sep=",")
df.head()
```

```python
df.plot(kind='scatter', x='Population', y='Profit', c='Red')
```

```python
x = list(df['Population'])
y = list(df['Profit'])
LinearRegression = []

# Defining b0 anda b1 initial values
#b0 = random.uniform(0,2)
#b1 = random.uniform(0,2)
initial_b0 = 5
initial_b1 = 5
b0 = initial_b0
b1 = initial_b1
print("Valores Iniciais de b0 e b1: ", b0, b1)

# Generating First Linear Regression Prediction
for i in range (len(x)):
    LinearRegression.append(lr.prediction(b0, b1, x[i]))

# Plotting First Linear Regression Prediction
df.plot(kind='scatter', x='Population', y='Profit', c='Red')
plt.plot(x, LinearRegression, color = 'g')
plt.title('Epoch: 0')
plt.show()
```

```python
c = lr.cost(b0, b1, x, y)
print("Erro Inicial: ", c)
```

```python
CostValues = []
alpha = 0.01
epoch = 0

# Initializing b0 and b1
b0 = initial_b0
b1 = initial_b1

# Running the model for 1000 epochs
while epoch <= 200:

    PredictionValues = []

    # Predict with the Linear Regression Model
    for i in range (len(x)):
        PredictionValues.append(lr.prediction(b0, b1, x[i]))
        
    # Calculate the cost of the prediction
    c = lr.cost(b0, b1, x, y)
    CostValues.append(c)

    # Calculate b0 and b1 adjustments values
    b0_adjust = lr.adjust_b0(b0, b1, x, y)
    b1_adjust = lr.adjust_b1(b0, b1, x, y)

    # Update b0 and b1 values
    temp_b0 = b0 - alpha * b0_adjust
    temp_b1 = b1 - alpha * b1_adjust
    b0 = temp_b0
    b1 = temp_b1

    # Plotting Linear Regression for each 500 epoch
    if (epoch == 0 or epoch == 100 or epoch == 200):
        df.plot(kind='scatter', x='Population', y='Profit', c='Red')
        plt.plot(x, PredictionValues, label = '', color = 'g')
        plt.title('Epoch: {}'.format(epoch))
        plt.show()

    # Incrementing the epoch
    epoch = epoch + 1

# Plotting Cost
x_cost = range(0,201)
t = list(x_cost)
plt.scatter(t, CostValues, color = 'Blue')
plt.title('Cost')
plt.show()
        
```

Conclusão:

Podemos notar com os gráficos acima que os ajustes levam a um modelo de regressão linear cada vez melhor com o passar das épocas.
Podemos notar também que a função custo decai muito rapidamente nas primeiras épocas, isso é devido ao alpha utilizado para fazer o ajuste, caso o alpha fosse menor, esse decaimento seria mais dividido entre as épocas, porém o importante é analisar que o erro tende a zero como esperado.
